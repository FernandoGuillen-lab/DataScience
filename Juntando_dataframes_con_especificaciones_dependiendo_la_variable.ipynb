```python
import pandas as pd
#Crea cuatro DataFrames en pandas, cada uno para cargar el contenido de los diferentes archivos.


test=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Test.csv")
test
```

```python
test2=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Test2.csv")
test2
```

```python
train=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Train.csv")
train
```

```python
train2=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Train2.csv")
train2
```

```python
#Conjunta de manera correcta la información, de tal forma que queden todos los registros juntos en un solo DataFrame que contenga todas las columnas necesarias.

df=pd.merge(test,test2,on=["Employee_ID"],how="outer")

df
```

```python
df1=pd.merge(train,train2,on=["Employee_ID"],how="outer")

df1
```

```python
df2=pd.concat([df,df1], axis=0,ignore_index=True)

df2
```

```python
#Para todos los registros que no tienen un valor para un sistema operativo, ya sea de su computadora o de su dispositivo móvil, asignar el valor “no registrado"

copia=df2.copy()
copia[" Computer_OS"].fillna("no registrado",inplace=True)
copia[" Mobile_OS"].fillna("no registrado",inplace=True)

copia
```

```python
#El porcentaje de personas que usan Windows / Mac / Linux / no registrado como su sistema operativo en su computadora.

porc1=copia[" Computer_OS"].value_counts(normalize=True)*100

porc1
```

```python
#El porcentaje de personas que usan iOS / Android / No registrado como su sistema operativo en su dispositivo móvil.

porc2=copia[" Mobile_OS"].value_counts(normalize=True)*100

porc2
```

```python
#La edad promedio de la gente que utiliza cada sistema operativo de escritorio.

prom1=copia.groupby(" Computer_OS")["Age"].mean()

prom1
```

```python
#El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio.

prom2=copia.groupby(" Computer_OS")["Education_Level"].mean()

prom2
```

```python
#El número total de tickets generados por la gente con cada sistema operativo de escritorio.

tic=copia.groupby(" Computer_OS")[" Computer_tickets"].sum()

tic
```

```python
#. Envía el resultado de cada solicitud de información a un archivo XLS[X], donde el resultado de cada consulta debe ir en su propia hoja de trabajo.

porc1ex=pd.DataFrame(porc1)
porc2ex=pd.DataFrame(porc2)
prom1ex=pd.DataFrame(prom1)
prom2ex=pd.DataFrame(prom2)
ticex=pd.DataFrame(tic)

with pd.ExcelWriter("resultados.xlsx") as writer:
  porc1ex.to_excel(writer,sheet_name="porc1")
  porc2ex.to_excel(writer,sheet_name="porc2")
  prom1ex.to_excel(writer,sheet_name="prom1")
  prom2ex.to_excel(writer,sheet_name="prom2")
  ticex.to_excel(writer,sheet_name="tic")
```
