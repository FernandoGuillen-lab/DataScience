```python
# Convertir las columnas de fecha al formato datetime
timco['E.T.D'] = timco['E.T.D'].str.strip()
timco['E.T.A'] = timco['E.T.A'].str.strip()
timco['E.T.D'] = pd.to_datetime(timco['E.T.D'], format='%d/%m/%Y', errors='coerce')
timco['E.T.A'] = pd.to_datetime(timco['E.T.A'], format='%d/%m/%Y', errors='coerce')

# Asegurarse de que la columna "Cantidad de Dias de Travesía" es de tipo numérico
timco['Cantidad de Dias de Travesía'] = pd.to_numeric(timco['Cantidad de Dias de Travesía'], errors='coerce')

```

```python
# Limpiar y convertir columnas de cantidades y peso
timco['Cantidad de cajas en contenedor'] = timco['Cantidad de cajas en contenedor'].astype(str).str.replace(',', '.')
timco['Peso bruto en Kg.'] = timco['Peso bruto en Kg.'].astype(str).str.replace(',', '.')
timco['Cantidad de cajas en contenedor'] = timco['Cantidad de cajas en contenedor'].str.replace('.', '', n=1).astype(float)
timco['Peso bruto en Kg.'] = timco['Peso bruto en Kg.'].str.replace('.', '', n=1).astype(float)
timco['Volumen en M3'] = timco['Volumen en M3'].astype(str).str.replace(',', '.').str.replace('.', '', n=1).astype(float)
```

```python
#Duración real en días
timco['Duración_real'] = (timco['E.T.A'] - timco['E.T.D']).dt.days

# Calcular la diferencia entre la duración real y la reportada
timco['Discrepancia_dias'] = timco['Duración_real'] - timco['Cantidad de Dias de Travesía']

#Se crea variable atrasos
timco["atrasos"]= timco["Discrepancia_dias"].apply(lambda x: 1 if x > 0 else 0)
```

## **ETAPA III:** Preparación de los datos

En esta etapa se aplican operaciones comunes para convertir los datos crudos del mundo real, en un conjunto de variables útiles para el aprendizaje automático.

El procesamiento puede incluir:
* Manejo de valores faltantes y atípicos
* Generación de nuevas características
* Discretización o binning
* Codificación
* Escalamiento
* Transformación

```python
#Reemplazo de valores faltantes

timco["Naviera"].fillna("Unknown", inplace= True)
timco["Estatus de embarque"].fillna("Pendiente", inplace= True)
timco["Transporte Terrestre"].fillna("Sin ID", inplace= True)
timco["Tipo de contenedor"].fillna("Desconocido", inplace= True)
timco["Tipo de carga"].fillna("Sin ID", inplace= True)

timco["Cantidad de Dias de Travesía"].fillna(0, inplace= True)
timco["Discrepancia_dias"].fillna(0, inplace= True)

#Eliminar variables para evitar muticolinealidad

timco.drop("Duración_real", axis= 1, inplace= True )
```

```python
t1= timco.copy()

categorica_ordinal = ["Estatus de embarque"]
categorica_nominal = ["Naviera", "Agente", "Embarcador", "Consignatario", "P.O.", "P.O.L", "P.O.D", "Transporte Terrestre", "Tipo de carga", "Tipo de contenedor", "Destino", "Descripción de la mercancia", "Buque"]
numerica = ["Cantidad de cajas en contenedor", "Peso bruto en Kg.", "Volumen en M3", "Cantidad de Dias de Travesía"]
#,  "Discrepancia_dias"
```

```python
#Transformar variables categóricas ordinales

variables_ordinales= t1[categorica_ordinal]
ordinal_encoder= OrdinalEncoder().fit_transform(variables_ordinales)
ordinal_df= pd.DataFrame(ordinal_encoder, columns= categorica_ordinal)

```

```python
#Transformar variables categóricas nominales

variables_nominales= t1[categorica_nominal]
one_hot = OneHotEncoder(sparse_output= False)
oh= one_hot.fit_transform(variables_nominales)
nominales_df= pd.DataFrame(oh, columns= one_hot.get_feature_names_out(categorica_nominal))

```

```python
#Transformar variables numericas

variables_numericas= t1[numerica]
imputer= SimpleImputer(strategy= "mean")
imp= imputer.fit_transform(variables_numericas)
scaler= StandardScaler().fit_transform(imp)
numericas_df= pd.DataFrame(scaler, columns= numerica)
```

```python
#Juntar los 3 dataframes creados en uno y añadir la variable de salida

t1_final= pd.concat([ordinal_df, nominales_df, numericas_df], axis= 1)
t1_final["atrasos"]= t1["atrasos"].values

```

```python
#Preparación de los datos

X= t1_final.drop("atrasos", axis=1)
y= t1_final["atrasos"]

x_train, x_test, y_train, y_test = train_test_split(X,y, stratify= y, test_size= .2, random_state= 42)

smote= SMOTE(sampling_strategy= .75)

x_train_smote, y_train_smote= smote.fit_resample(x_train, y_train)
```

```python
#Primer modelo LogisticRegression

clf_lr= LogisticRegression(solver= "liblinear",  random_state= 42, class_weight= "balanced")

params_lr = {
    'C':[.01, .1, 1, 10, 100],
    'penalty':["l1", "l2"]
}

grid_search_lr= GridSearchCV( clf_lr, scoring= "roc_auc", param_grid= params_lr, cv= 3, return_train_score= True, n_jobs= -1)

grid_search_lr.fit(x_test, y_test)
```

```python
def resumen_grid (df):

  best_params= grid_search_lr.best_params_

  print("\n Se evaluaron {} modelos utilizando el grid search.".format(pd.DataFrame(grid_search_lr.cv_results_).shape[0]))
  print("\n Los hiperparametros del mejor modelo son:" + str(best_params) + "\n")
```
